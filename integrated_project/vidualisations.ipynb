# Vidual represation of the collected insurance claims data.
### This should give a proper base for the Data Science team as they start with they EDA

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df  = pd.read_csv('cleaned_insurance_claims_data.csv')
```

## Distribution of Policy Annual Premium

The histogram bellow shows the **distribution of annual premiums** paid by policyholders. The data indicates that most premiums fall between **\400 and \2000** with the highest premium rate between **\1200 and \1400. 

The presence of a peak at lower premium values suggests that a significant number of customers prefer policies with lower annual premiums, possibly indicating price sensitivity.


```python
# Distribution of Policy Annual Premium
plt.figure(figsize=(8, 6))
sns.histplot(df['policy_annual_premium'], kde=True, bins=10, color='blue')
plt.title('Distribution of Policy Annual Premium')
plt.xlabel('Policy Annual Premium')
plt.ylabel('Frequency')
plt.show()
```

## Age Group vs. Total Claim Amount

This bar chart illustrates how **total claim amounts vary by age group**. Customers aged **50-65** report the highest average total claim amounts, while those in the **30-50** and **<30** age groups have the lowest average claims.

This trend could reflect differing risk profiles or coverage preferences across age groups. For example, older customers may opt for policies with greater coverage, leading to higher claim amounts.

```python
# Age Group vs. Total Claim Amount
plt.figure(figsize=(8, 6))
sns.barplot(x='age_group', y='total_claim_amount', data=df, palette='viridis')
plt.title('Age Group vs. Total Claim Amount')
plt.xlabel('Age Group')
plt.ylabel('Average Total Claim Amount')
plt.show()
```

## Tenure Bracket vs. Number of Claims

The countplot shows the distribution of claims across **tenure brackets** (how long a customer has been with the company). Most claims come from customers in the **>10 years bracket**, with fewer claims from short-tenured customers.

```python
# Tenure Bracket vs. Number of Claims
plt.figure(figsize=(8, 6))
sns.countplot(x='tenure_bracket', data=df, palette='magma')
plt.title('Tenure Bracket vs. Number of Claims')
plt.xlabel('Tenure Bracket')
plt.ylabel('Count of Claims')
plt.show()
```

## Policy Deductible vs. Total Claim Amount

The boxplot below examines the relationship between **policy deductibles** and **total claim amounts**. 

### Deductible 500:

Has the largest spread (range of data points) and a higher median compared to the others.
Indicates claim amounts for this deductible tend to be higher and more variable.

### Deductible 2000:

Has smaller IQR, lower median, and visible outliers below the box.
Suggests most claim amounts are smaller for this deductible, but a few policies generate outliers.

### Deductible 1000:

Shows a moderate spread and is less variable compared to 500.

```python
# Policy Deductible vs. Total Claim Amount
plt.figure(figsize=(8, 6))
sns.boxplot(x='policy_deductable', y='total_claim_amount', data=df, palette='coolwarm')
plt.title('Policy Deductible vs. Total Claim Amount')
plt.xlabel('Policy Deductible')
plt.ylabel('Total Claim Amount')
plt.show()
```

## Fraud Reported by Incident Severity

The stacked bar chart below shows the **relationship between incident severity and fraud reporting**. Fraud is more commonly reported for **major and severe incidents**, while **minor incidents** rarely involve fraud.

This pattern could be due to the larger financial incentives to commit fraud in high-severity cases or stricter investigations for such incidents.

```python
# Fraud Reported by Incident Severity
fraud_severity_counts = df.groupby(['incident_severity', 'fraud_reported']).size().reset_index(name='count')
fig = px.bar(fraud_severity_counts, x='incident_severity', y='count', color='fraud_reported', barmode='stack',
             title='Fraud Reported by Incident Severity')
fig.show()
```

## Number of Bodily Injuries vs. Total Claim Amount

The scatterplot below examines the relationship between **bodily injuries sustained** and **total claim amounts**. 

As expected, higher bodily injury counts are associated with larger claim amounts. Fraud is more frequently reported in these cases, as indicated by the distinction in the color coding.

This insight could help flag potentially fraudulent claims for further investigation.

```python
# Number of Bodily Injuries and Claim Amount
plt.figure(figsize=(8, 6))
sns.scatterplot(x='bodily_injuries', y='total_claim_amount', data=df, hue='fraud_reported', palette='deep', s=100)
plt.title('Number of Bodily Injuries vs. Total Claim Amount')
plt.xlabel('Number of Bodily Injuries')
plt.ylabel('Total Claim Amount')
plt.legend(title='Fraud Reported')
plt.show()
```
