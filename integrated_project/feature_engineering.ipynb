```python
import pandas as pd
```

```python
df  = pd.read_csv('cleaned_insurance_claims_data.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
# Define tenure bins and labels
tenure_bins = [0, 12, 36, 60, 120, float('inf')]  # Bins: 0-12, 13-36, etc.
tenure_labels = ['<1 year', '1-3 years', '3-5 years', '5-10 years', '>10 years']

# Create the 'tenure_bracket' column
df['tenure_bracket'] = pd.cut(df['months_as_customer'], bins=tenure_bins, labels=tenure_labels, right=False)

df.head(10)
```

```python
# Define age bins and labels
age_bins = [0, 30, 50, 65, float('inf')]  # Bins: <30, 30-50, 50-65, >65
age_labels = ['<30', '30-50', '50-65', '>65']

# Create the 'age_group' column
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)

df.head(10)
```

```python
df.to_csv('Cleaned_insurance_claims_data.csv', index=False)
```
